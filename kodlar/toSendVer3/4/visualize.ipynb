```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = "data.csv"
df = pd.read_csv(file)
df.head()
```

```python
x_axis = df['x']
y_axis = df['y']

x_axis = list(x_axis)
y_axis = list(y_axis)
```

```python
plt.scatter(x_axis, y_axis)
plt.xlabel("x -> ")
plt.ylabel("y -> ")
plt.title("data")
plt.show()
```

```python
sns.jointplot(data=df, x="x", y="y")
```
